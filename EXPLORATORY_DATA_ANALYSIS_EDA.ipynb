```python
# Importing libraries
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
# Loading csv files
df = pd.read_csv("selected_viviendas.csv")  #  selected_concentradohogar   selected_viviendas
```

```python
# Generating the profiling report
profile = ProfileReport(df, title="Pandas Profiling Report - selected_viviendas", explorative=True)
```
